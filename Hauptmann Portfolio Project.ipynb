```python
class Book:
    def __init__(self, title, author, available):
        self.title = title
        self.author = author
        self.available = available
    
    def display_info(self):
        print(f"BOOK INFORMATION:\nTitle: {self.title}\nAuthor: {self.author}\nAvailable: {str(self.available)}\n\n")
    
    def borrow_book(self, book):
        if book.available:
            book.available = False
            
    def return_book(self, book):
        if not book.available:
            book.available = True
            
class User:
    def __init__(self, name, books):
        self.name = name
        self.books = books
        
    def display_books(self):
        print("BOOKS CHECKED OUT:\n")
        n = 1
        for i in self.books:
            print(f"{n}: {i.title}")
            n += 1
        print("\n")
        
    def borrow_book(self, book):
        self.books.append(book)
        
    def return_book(self, book):
        self.books.remove(book)
        
class Library:
    def __init__(self, books, users):
        self.books = books
        self.users = users
    
    def add_book(self, book):
        self.books.append(book)
        
    def add_user(self, user):
        self.users.append(user)
        
    def display_books(self):
        print("AVAILABLE BOOKS:")
        n = 1
        for i in self.books:
            print(f"{n}: {i.title}")
            n += 1
        print("\n")
    
    def display_users(self):
        print("USERS:")
        for i in self.users:
            print(i.name)
        print("\n")
            
library = Library([], [])
library.add_book(Book("The Hitchhiker's Guide to the Galaxy", "Douglas Adams", True))
library.add_book(Book("Musicophilia", "Oliver Sacks", True))
library.add_book(Book("Got Yer Brain!", "Dr. Ben Carson", True))

user_name = input("Welcome to the Python Library! Enter your name: ")
user = User(user_name, [])
library.add_user(user)

nav = 0
while nav < 6:
    print("Please select an option:\n1: Browse available books\n2: Check out a book\n3: Return a book\n4: Change user\n5: Show list of users\n6: Log out")
    try:
        nav = int(input(">"))
    except ValueError:
            print("ERROR: Input must be an integer\n")
    if nav == 1:
        library.display_books()
        try:
            book_index = int(input("Enter a book number for more information or 0 to return to main menu: "))
        except ValueError:
            print("ERROR: Input must be an integer\n")
        if book_index == 0:
            pass
        else:
            try:
                book = library.books[book_index - 1]
            except IndexError:
                print("ERROR: Input out of range\n")
            else:
                book.display_info()
    if nav == 2:
        library.display_books()
        try:
            book_index = int(input("Enter the number of the book to check out or 0 to return to main menu: "))
        except ValueError:
            print("ERROR: Input must be an integer\n")
        if book_index == 0:
            pass
        else:
            try:
                book = library.books[book_index - 1]
            except IndexError:
                print("ERROR: Input out of range\n")
            else:
                if book.available:
                    for x in (user, book):
                        x.borrow_book(book)
                    print(f"You have checked out {book.title}. Thank you!\n")
                else:
                    print("Book not available\n")
    if nav == 3:
        user.display_books()
        try:
            book_index = int(input("Enter the number of the book to return or 0 to return to main menu: "))
        except ValueError:
            print("ERROR: Input must be an integer\n")
        if book_index == 0:
            pass
        else:
            try:
                book = user.books[book_index - 1]
            except IndexError:
                print("ERROR: Input out of range\n")
            else:
                if not book.available:
                    for x in (user, book):
                        x.return_book(book)
                    print(f"You have returned {book.title}. Thank you!\n")
    if nav == 4:
        user_name = input("Enter name of new user or 0 to return to main menu: ")
        if user_name == 0:
            pass
        else:
            for u in library.users:
                if u.name == str(user_name):
                    user = u
                    break
                else:
                    user = User(user_name, [])
                    library.add_user(user)
    if nav == 5:
        library.display_users()
    if nav > 6 or nav < 0:
        nav = 0
```
